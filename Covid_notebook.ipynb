```python
-- Select data to explore
SELECT 
	location,
	date,
	total_cases,
	new_cases,
	total_deaths,
	population
FROM 'CovidDeaths.csv'
ORDER BY 1,2
LIMIT 10;
```

```python
-- Total Deaths Vs Total Cases
SELECT 
	location,
	date,
	total_cases,
	total_deaths,
	(total_deaths/total_cases) * 100 as percent
FROM 'CovidDeaths.csv'
WHERE location LIKE '%States%'
ORDER BY 1,2
LIMIT 10;
--Shows the likelyhood of dieing from covid started decreasing starting in September of 2020 
```

```python
-- Percentage of population who have experienced Covid over time
SELECT 
	location,
	date,
	total_cases,
	population,
	(total_cases/population) * 100 as percent
FROM 'CovidDeaths.csv'
WHERE location LIKE '%States%'
ORDER BY 1,2
--LIMIT 10;
```

```python
-- Counties with the highest death counts

SELECT
	location,
	max(total_deaths) as total_death_count
FROM 'CovidDeaths.csv'
WHERE continent IS NOT null
GROUP BY location, population
ORDER BY total_death_count desc
limit 10
```

```python
-- Break down by Continent

SELECT
	location,
	max(total_deaths) as total_death_count,
	round(total_death_count/population * 100,4) as percent
FROM 'CovidDeaths.csv'
WHERE continent IS null
	AND location <> 'World'
	AND location <> 'International'
GROUP BY location,population
ORDER BY total_death_count desc
limit 10
```

```python
-- Join Vaccine data and find percentage of the population who is vaccinated
SELECT 
	d.location,
	round(max(people_fully_vaccinated)/d.population *100,4) as percent_vaccinated
FROM 'CovidDeaths.csv' as d
JOIN 'CovidVaccinations.csv' as v
	ON d.location = v.location
	AND d.date = v.date
WHERE d.continent IS NOT null
GROUP BY d.location, d.population
ORDER BY percent_vaccinated DESC
limit 10
```
